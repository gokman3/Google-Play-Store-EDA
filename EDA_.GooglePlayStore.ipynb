6.	Handle incorrect data types.

```python
#10472 index data is shifted to the left. I will drop this data because it will still be nan.
data.drop(index=10472, inplace=True)
data.reset_index(drop=True, inplace=True)
```

```python
#Chancing the reviews columns data type object to Int
data["Reviews"] = pd.to_numeric(data["Reviews"]).astype("Int64")
```

```python
#Chancing the Size column data type object to float type
#Size column has M or k string so we should clean them because if they have a string value we can t change the type to float

def sizeChanger(size):
    if type(size) == float or type(size) == int:
        return float(size)
    elif "Varies in device" in size:
        return float(size.replace("Varies in device", np.nan))
    elif 'M' in size:
        return float(size.replace('M', ''))
    elif 'k' in size:
        return float(size.replace('k', ''))/1024
    elif pd.isna(size):
        return np.nan
    else:
        return np.nan
data["Size"] = data["Size"].apply(sizeChanger)
data["Size"]= pd.to_numeric(data["Size"]).astype("float")
```

```python
#Chancing the Installs column data type object to float type
#Install column has "+"and ","  strings in values so we should clean them.
def installChanger(installs):
    if type(installs) == float or type(installs) == int:
        return float(installs)
    elif "+" and "," in installs:
        newInstalls = installs.replace("+", "")
        newInstalls = newInstalls.replace(",", "")
        return float(newInstalls)
    elif "+" in installs:
        return float(installs.replace("+", ""))
    elif "," in installs:
        return float(installs.replace(",", ""))
    elif pd.isna(installs):
        return float(np.nan)
    else:
        return float(np.nan)

data["Installs"] = data["Installs"].apply(installChanger)
data["Installs"]= pd.to_numeric(data["Installs"]).astype("float")
```

```python
#Chancing the Price column data type object to float
#Price column has "$" string in values so we should clean them
def priceChanger(price):
    if type(price) == str:
        if "$" in price:
            return float(price.replace("$", ""))
        else:
            return float(price)
    elif type(price)==float or type(price)==int :
        return float(price)
    elif pd.isna(price):
        return float(np.nan)
    else:
        return float(np.nan)

data["Price"] = data["Price"].apply(priceChanger)
data["Price"]= pd.to_numeric(data["Price"]).astype("float")
```

```python
#Chancing the Last Updated column data type object to datetime
data['Last Updated'] = pd.to_datetime(data['Last Updated'])
```
